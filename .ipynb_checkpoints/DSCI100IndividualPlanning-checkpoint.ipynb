

1) The dataset that we have chosen is the players.csv dataset. This dataset contains 9 variables (columns) and 196 observations (rows).

Potential Issues
One visible issue is the data is the amount of NA values for organizationname and individualID in the dataset. Along with this, from the previewed data, there seems to be a large amount of users who have zero played_hours, which could cause problems when using that as a predictor. It is also important to ensure that we have a similar amount of observations that are scaled and centered for all predictors, of we are to use either k-nn classification or k-nn regression


2) The question we will answer is, How does previous experience with Minecraft influence users' engagement with the game? For this, we will use the experienced and played_hours columns as predictors. These will help us address our question because the experience is self-reported by the user as what level they believe they are at before entering the server, therefore we can get a decent idea of the amount of previous experience they have with the game. We can use the played_hours column to see how much time a user spent engaged in the server. Then we can attempt to see a relationship between these two and report our findings.

3) Main insights from data below
- Many players with 0 play time
- Many players with age around 20
- Few users with over 200 hours played


```r
library(tidyverse)

players <- read_csv("players.csv")
players

#sessions <- read_csv("sessions.csv")

players_histo <- ggplot(players, aes(x = played_hours)) +
       geom_histogram() +
       labs(x = "Player Hours of users")

players_histo


players_scatter <- ggplot(players, aes(x = played_hours, y = age, color = experience)) +
                          geom_point() +
                          labs(x = "Played hours for user", y = "Age of user", color = "Experience of user")

players_scatter


players_box <- ggplot(players, aes(x = experience, y = played_hours)) +
    geom_boxplot()

players_box
```

4)  We are going to use k-nn regression to answer our question. This is appropriate because we can uncover general trends using forms of regression, however, since we don’t know about the complexity of the relationship or if it is linear, k-nn regression is a safer option than linear regression. We will assume that all people with 0 hours played are not a part of this study as they may have not had a chance to play yet. Some potential limitations of this may be choosing the wrong number of k-values, which may cause either overfitting or underfitting in the visualization of the data. We will split the data into 70-15-15 (training-testing-validation) using the intitial_split function. We will split the data once at the beginning of our preprocessing data so that we can effectively scale and center our training data. We will use the validation set to tune our model, prevent overfitting, and sense when the quality of our data begins to deteriorate. Finally we will use cross-validation to pick the number of K values with lowest Root Mean Squared Prediction Error (RMSPE).
